```python
# import pandas as pd
import pandas as pd
```

```python
df=pd.read_csv("/content/banglor house data use support vector regression.csv")
```

```python
df.shape
```

# split data

```python
x=df.drop('price',axis=1) # drop dependent future
y=df['price']
x.shape , y.shape
```

train test split data

```python
#train_test split jdata
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# SUPPORT VECTOOR REGRESSION

```python
from sklearn.svm import SVR
```

```python
SR=SVR(kernel="rbf" ) # by default kernal
SR.fit(x_train,y_train)
```

```python
SR.score(x_test,y_test)
```

```python
SR=SVR(kernel="linear" ) #check linear kernal
SR.fit(x_train,y_train)
```

```python
SR.score(x_test,y_test)  #score
```

```python
y_ped=SR.predict(x_test) # predict
y_ped
```

```python
y_test
```

root mean squaare


```python
#finnd mean squre error
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse=mean_squared_error(y_test,SR.predict(x_test))
rmse=np.sqrt(mse)
```
