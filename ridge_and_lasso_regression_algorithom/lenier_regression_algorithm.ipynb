```python
# drop dependent varriable
x=df.drop('price',axis=1)
y=df['price']
```

# train_test_split_dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

# standard_scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

# lenier regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_  # check coficient
```

```python
lr.intercept_ # check intercept
```

```python
x_test[0,:] # single house ki value
```

```python
lr.predict([x_test[0,:]]) # single house ki value ko predict
```

```python
lr.predict(x_test) # predict multiple house price
```

```python
# check accuracy
lr.score(x_test,y_test)
```

# *ridge_and_lasso_algo*

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rd=Ridge() # by dafault value of alpha=1
rd.fit(x_train,y_train)
# find score
rd.score(x_test,y_test)
```

# LASSO ALGO

```python
ls=Lasso() # by default value alpha=1
ls.fit(x_train,y_train)
# find score
ls.score(x_test,y_test)
```

```python
rd=Ridge(alpha=2) # encrease value of alpha=1
rd.fit(x_train,y_train)
# find score
rd.score(x_test,y_test)
```

```python
ls=Lasso(alpha=2) # encrease value of alpha=1# by default value alpha=1
ls.fit(x_train,y_train)
# find score
ls.score(x_test,y_test)
```

```python
ls=Lasso(alpha=3) # encrease value of alpha=1# by default value alpha=1
ls.fit(x_train,y_train)
# find score
ls.score(x_test,y_test)
```
