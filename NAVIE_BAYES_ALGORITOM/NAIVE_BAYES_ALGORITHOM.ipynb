```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import datasets
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data=load_breast_cancer()
```

```python
cancer_data.data
```

```python
cancer_data.feature_names
```

```python
cancer_data.target
```

```python
cancer_data.target_names
```

```python
# conver datafeame
df=pd.DataFrame(np.c_[cancer_data.data,cancer_data.target],columns=[np.append(cancer_data.feature_names,["target"])])
```

```python
df.head()
```

```python
x=df.drop("target",axis=1)
y=df["target"]
```

```python
x.shape
```

```python
y.shape
```

# split_train_test_data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=202)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# future scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)

#
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df,annot=True, linecolor="b",linewidths=2)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True  , linecolor="k",linewidths=2)
```

# import navie bays

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gauss=GaussianNB()
gauss.fit(x_train,y_train)
gauss.score(x_test,y_test)
```
